```python
import tensorflow as tf
from keras import datasets,models,layers,losses,optimizers
```

```python
(x_train, y_train), (x_test, y_test) = datasets.fashion_mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
cnn = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPool2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPool2D((2,2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10)
])
cnn.compile(optimizer='adam', loss=losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
cnn.summary()
```

```python
cnn.fit(x_train, y_train, epochs=10, batch_size=64, verbose=2)
```

```python
cnn_loss, cnn_accuracy = cnn.evaluate(x_test, y_test, verbose=1)
print(f'CNN Accuracy on test set: {cnn_accuracy}')
```

MLP models accuracy is around 88% whereas cnn models accuracy is around 90%.
